Groupby and aggregating

```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\abstr\OneDrive\Desktop\PYTHON\Pandas\Flavors.csv")
```

```python
df
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame.mean(numeric_only=True)
```

```python
df.groupby('Base Flavor').mean(numeric_only=True)
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').sum(numeric_only=True)
```

agg() aggregate func, we need to provide a dictionary

```python
df.groupby('Base Flavor').agg({'Flavor Rating':['mean','max','count','sum']})
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating':['mean','max','count','sum'],'Texture Rating':['mean','max','count','sum']})
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','max','count','sum']})
```

Merge, Join and Concatenate

```python
#Combining two seperate dataframes
```

```python
df1=pd.read_csv(r"C:\Users\abstr\OneDrive\Desktop\PYTHON\Pandas\LOTR.csv")
df1
```

```python
df2=pd.read_csv(r"C:\Users\abstr\OneDrive\Desktop\PYTHON\Pandas\LOTR 2.csv")
df2
```

```python
#merge
```

```python
df1.merge(df2)

#by default this performs inner join
```

```python
df1.merge(df2,how='inner')
```

```python
df1.merge(df2,how='inner',on='FellowshipID')
```

```python
df1.merge(df2,how='inner',on=['FellowshipID','FirstName'])
```

```python
#lets see outer join
```

```python
df1.merge(df2,how='outer')

#it gives everything plus the missing values as NaN
```

```python
df1.merge(df2,how='left')

#Gives full df1 and the common part with df2
```

```python
df1.merge(df2,how='right')
```

```python
df1.merge(df2,how='cross')
```

Joins

For joins, the dataframes join on indexes, we have to specify which index are we joining on.

```python
df4=df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix='_left', rsuffix = '_Right',how='outer')
df4
```

Concatenate

```python
#first data frame sits on top of the other, there is no joining or merging the dataframe.
```

```python
pd.concat([df1,df2])
```
